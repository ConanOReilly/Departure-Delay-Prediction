# **Finding Aiports With Most Entries**
Selected the top two airports with the most entries for testing the predictive model.

```python
import pandas as pd

dataset_path = 'balanced_dataset111.csv'
df = pd.read_csv(dataset_path)

print("Displaying the first 5 rows of the dataset:")
df.head()
```

```python
# counting the occurrences of each airport in the 'DEPARTING_AIRPORT' column
airport_counts = df['DEPARTING_AIRPORT'].value_counts()

# finding the airport with the highest number of occurrences
most_common_airport = airport_counts.idxmax()
most_common_count = airport_counts.max()

print(f"The airport with the highest number of rows (occurrences) is '{most_common_airport}' with {most_common_count} flights.")

print("\nAll airports sorted by number of occurrences:")
print(airport_counts)
```

```python
# filtering data for 'Atlanta Municipal'
atlanta_df = df[df['DEPARTING_AIRPORT'] == 'Atlanta Municipal']

# filtering data for 'Chicago O'Hare International'
chicago_df = df[df['DEPARTING_AIRPORT'] == "Chicago O'Hare International"]

# saving the filtered datasets to separate CSV files
atlanta_df.to_csv('atlanta_municipal_dataset.csv', index=False)
chicago_df.to_csv('chicago_ohare_dataset.csv', index=False)

print("Filtered datasets created and saved as 'atlanta_municipal_dataset.csv' and 'chicago_ohare_dataset.csv'.")
```
