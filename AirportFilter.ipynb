# **Airport Filter**
The following code filters the dataset for entries from specific airport for a more narrower analysis.

```python
import pandas as pd

dataset_path = 'balanced_dataset111.csv'
df = pd.read_csv(dataset_path)

print("Displaying the first 5 rows of the dataset:")
df.head()
```

```python
# extracting and displaying unique airport names
print("List of all unique airport names:")
airport_names = df['DEPARTING_AIRPORT'].unique()
print(airport_names)

print("\nTotal number of unique airports:", len(airport_names))
```

```python
# filtering the dataset to only keep rows where DEPARTING_AIRPORT is 'John F. Kennedy International'
filtered_df = df[df['DEPARTING_AIRPORT'] == 'John F. Kennedy International']

# displaying the filtered dataset
print("Filtered dataset with only 'John F. Kennedy International':")
print(filtered_df)

filtered_df.to_csv('filtered_dataset.csv', index=False)
```
